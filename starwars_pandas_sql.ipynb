### Data Extraction

#### Extraction of People data

```python
path = 'https://swapi.dev/api/people/?page={page}' # path to the API

next_page = True
page = 1
people_df = pd.DataFrame()

while next_page:
  df = pd.read_json(path.format(page=page)) # read the API
  if df['next'].isna().values[0]: # check if there is a next page
    next_page = False
  else:
    page += 1
  df_results = pd.DataFrame(df['results'].values.tolist()) # convert the results to a dataframe
  people_df = pd.concat([people_df, df_results], ignore_index=True) # concat the results to the main dataframe
```

```python
path = 'https://swapi.dev/api/films/?page={page}'

next_page = True
page = 1
films_df = pd.DataFrame()

while next_page:
  df = pd.read_json(path.format(page=page))
  if df['next'].isna().values[0]:
    next_page = False
  else:
    page += 1
  df_results = pd.DataFrame(df['results'].values.tolist())
  films_df = pd.concat([films_df, df_results], ignore_index=True)
```

#### INNER JOIN

SQL query to display a comprehensive list of all the films needs to be provided, including their titles, directors, and release dates, where at least one character (people) from the film has also piloted a vehicle. Additionally, the names of these characters and the names of the vehicles they piloted in the respective films must be included.

```python
pd.read_sql(
  '''
    SELECT DISTINCT 
      f.title AS Film_Title, 
      f.director AS Director, 
      f.release_date AS Release_Date, 
      p.name AS Character_Name, 
      v.name AS Vehicle_Name
    FROM 
      films f
    INNER JOIN 
      people_films pf ON f.Id = pf.film_id
    INNER JOIN 
      people p ON pf.people_id = p.Id
    INNER JOIN 
      people_vehicles pv ON p.Id = pv.people_id
    INNER JOIN 
      vehicles v ON pv.vehicle_id = v.Id
    ORDER BY 
      f.title, p.name
  ''',
  con
)
```

#### LEFT JOIN

SQL query to display the names of the vehicles that have not been piloted but used in George Lucas's films between 1970 and 1980 along with films tile.

```python
pd.read_sql(
  '''
    SELECT 
      v.name AS Vehicle_Name, 
      f.title AS Film_Title,
      f.release_date AS Release_Date
    FROM 
      vehicles v
    LEFT JOIN 
      films_vehicles fv ON v.Id = fv.vehicle_id
    LEFT JOIN 
      people_vehicles pv ON v.Id = pv.vehicle_id
    LEFT JOIN 
      films f ON fv.film_id = f.Id
    WHERE 
      pv.people_id IS NULL AND f.director = 'George Lucas' AND f.release_date BETWEEN '1970-01-01' AND '1980-01-01'
    ORDER BY 
      v.name, f.title
  ''',
  con
)
```

#### GROUP BY

SQL query to display the distribution of film budgets (in credits) based on the gender of the characters involved in each film, and which film had the highest budget for male and female characters respectively.

```python
pd.read_sql(
  '''
    SELECT 
      f.title AS film_title,
      p.gender,
      SUM(v.cost_in_credits) AS total_budget
    FROM 
      films AS f
    INNER JOIN
      people_films AS pf ON f.Id = pf.film_id
    INNER JOIN 
      people AS p ON pf.people_id = p.Id
    INNER JOIN 
      people_vehicles AS pv ON p.Id = pv.people_id
    INNER JOIN 
      vehicles AS v ON pv.vehicle_id = v.Id
    GROUP BY 
      f.title, p.gender
    ORDER BY 
      f.title, p.gender, total_budget DESC;
  ''',
  con
)
```

#### HAVING

SQL query to display the films that have characters with an average mass greater than 70, and have vehicles with a maximum atmosphering speed greater than 500 and to provide the film titles, directors, and the names of the vehicles that meet these criteria.

```python
pd.read_sql(
  '''
    SELECT 
      f.title AS Film_Title, 
      f.director AS Director, 
      v.name AS Vehicle_Name
    FROM 
      films f
    JOIN 
      people_films pf ON f.Id = pf.film_id
    JOIN 
      people p ON pf.people_id = p.Id
    JOIN 
      people_vehicles pv ON p.Id = pv.people_id
    JOIN 
      vehicles v ON pv.vehicle_id = v.Id
    GROUP BY 
      f.title, 
      f.director, 
      v.name
    HAVING 
      AVG(p.mass) > 70  AND MAX(v.max_atmosphering_speed) > 500
  ''',
  con
)
```

#### Common Table Expression (CTE)

SQL query to display which films had the most diverse cast in terms of gender representation among characters and to provide a breakdown of the number of male and female characters in each film, along with the total count. Also, to identify any films where there were no female characters at all.

```python
pd.read_sql(
  '''
    WITH CharacterGenderCounts AS (
      SELECT 
        pf.film_id,
        COUNT(CASE WHEN p.gender = 'male' THEN 1 END) AS male_count,
        COUNT(CASE WHEN p.gender = 'female' THEN 1 END) AS female_count
      FROM 
        people p
      INNER JOIN 
        people_films pf ON p.Id = pf.people_id
      GROUP BY 
        pf.film_id
    )

    SELECT 
      f.title AS Film_Title,
      f.release_date AS Release_Date,
      COALESCE(cgc.male_count, 0) AS Male_Characters,
      COALESCE(cgc.female_count, 0) AS Female_Characters
    FROM 
      films f
    LEFT JOIN 
      CharacterGenderCounts cgc ON f.Id = cgc.film_id
    WHERE 
      cgc.female_count > 0 OR cgc.female_count IS NULL
    ORDER BY 
      Female_Characters DESC, Male_Characters DESC, Film_Title
  ''',
  con
)
```
